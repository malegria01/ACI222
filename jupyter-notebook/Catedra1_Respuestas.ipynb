### Ejercicio 2 (5 puntos) 
Escribir un programa que pida al usuario ingresar un número entero, impar mayor o igual que 5 (>= 5) y que muestre por pantalla la siguiente figura. 

```
	*
    **
    ***
    ****
    *****
    ****
    ***
    **
    *
```

PD: Si no logra generar correctamente la figura, el puntaje es 0

````python
n = int(input("\nIngrese un numero entero impar mayor o igual que 5:\n"))
contain =''
for i in range(n+1):
    for j in range(i):
        contain +='*'
    if i != n+1:    
        contain +='\n'

for i in range(n):
    for j in range(n - i): 
        contain += '*'
    contain += '\n' 
print(contain)
````

    Escribir bien el programa que usa/llama a la función = 5 puntos



````python

#Function
def convertir_angulo(unidades=None, angulo_original=None):
    """Función que convierte de grados a radianes y vice-versa
    Parametros:
    unidades: unidades del angulo original: G para grados y R para radianes 
    angulo_original: Angulo a ser convertido.
    Ejemplo:
    convertir_angulo('G', 45)
    """
    pi =3.14159265358979323
    if unidades == 'G':
        #degrees to radians
        degree = angulo_original*(pi/180)
        return degree
    elif unidades == 'R':
        #radians to degrees
        radians = angulo_original*(180/pi)     
        return radians
    else:
        print("Valor invalido")
        help(angulo_original)


#short program to use the function 
        
angulo = float(input("Ingrese ángulo a convertir : "))
unidad = input("Ingrese G si su ángulo esta en grados y R si su ángulo esta en radianes: ")

resultado = convertir_angulo(unidades=unidad, angulo_original=angulo)

if unidad == 'G':
    print("\n{} grados is {:.4f} radianes".format(angulo,resultado))

elif unidad == 'R':
    print("\n{} radianes is {:.4f} grados".format(angulo,resultado))
````

````python
# calculadora

#FUNCIONES DE LA CALCULADORA

#FUNCION QUE SUMA DOS VALORES
def sumar(valor1=0,valor2=0):
    return valor1 + valor2 

#FUNCION QUE RESTA DOS VALORES
def restar(valor1=0,valor2=0):
    return valor1 - valor2 

#FUNCION QUE MULTIPLICACION DOS VALORES
def multiplicar(valor1=0,valor2=0):
    return valor1 * valor2 

#FUNCION QUE DIVIDE DOS VALORES
def dividir(valor1=0,valor2=0):
    return valor1 / valor2 


#FUNCIONES DE LA CALCULADORA
def menu():
    #MENU DE OPCIONES
    print("** ::::::::::::::::::::::::: **")
    print(":: Seleccione una operación ::")
    print("** ::::::::::::::::::::::::: **")
    print("-------------------------------")
    print("| Suma:             ->  1     |")
    print("| Resta:            ->  2     |")
    print("| Multiplicación:   ->  3     |")
    print("| División:         ->  4     |") 
    print("| Salir:            ->  5     |") 
    print("-------------------------------")
    print("** ::::::::::::::::::::::::: **")
    opcion = int(input())
    return opcion
    
opcion = menu()

while opcion < 1 or opcion >5:
    print("\nOpción no válida")
    print("------------------\n")
    opcion = menu()
    

if opcion == 5:
    print("Adios!")
else:
    n1 = int(input("\nIngrese numero 1:"))
    n2 = int(input("\nIngrese numero 2:"))

if opcion == 1:
    suma = sumar(n1,n2)
    print ("\nLa suma es de {}".format(suma))
elif opcion == 2:
    resta = restar(n1,n2)
    print ("\nLa resta es de {}".format(resta))
elif opcion == 3:
    multiplicar = multiplicar(n1,n2)
    print ("\nLa multipliccion es de {}".format(multiplicar))
elif opcion == 4:
    dividir = dividir(n1,n2)
    print ("\nLa división es de {}".format(dividir))
    


````
