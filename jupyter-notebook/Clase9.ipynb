## Data Frame

Un DataFrame es una estructura de datos con dos dimensiones en la cual se puede guardar datos de distintos tipos (como caractéres, enteros, valores de punto flotante, factores y más) en columnas. Es similar a una hoja de cálculo o una tabla de SQL. Un DataFrame siempre tiene un índice (con inicio en 0). El índice refiere a la posición de un elemento en la estructura de datos.

Si quieres saber más de lo que es un Data frame, siempre es mejor ir a la informacion oficial en la seccion [DataFrame](https://pandas.pydata.org/pandas-docs/stable/user_guide/dsintro.html)

En algunas ocasiones queremos calcular estadísticas de datos **agrupados** por **subconjuntos** o **atributos** de nuestros datos. Por ejemplo, si queremos calcular el promedio de peso de nuestros individuos por sitio.

Podemos calcular algunas estadísticas básica de todos los datos en una columna usando el siguiente comando:

````python
surveys_df['weight'].describe()
````

### La funcion ```describe()``` genera un dataframe con la media, la desviación estándar y los valores entre cuartiles de las columnas. Esta función excluye las columnas de texto y de fechas y sólo funciona con valores numéricos

````python
# Se puede obtener una métrica en particular:
cantidad = surveys_df['weight'].count()
mean = surveys_df['weight'].mean()
std = surveys_df['weight'].std()
minimo = surveys_df['weight'].min()
maximo = surveys_df['weight'].max()
q = surveys_df['weight'].quantile([0.10,0.5,0.9])
print(cantidad)
print(mean)
print(std)
print(minimo)
print(maximo)
print(q)
````

### Si ahora necesitamos extraer información por una o más variables se usa groupby


````python
#Datos agrupados por sexo
grouped_data = surveys_df.groupby('sex')
````

````python
# Estadísticas para todas las columnas numéricas por sexo
grouped_data.describe()
````

### Ejercicio de graficar:

Crear una gráfica de barras apiladas que muestre en el eje X cada uno de los sitios donde se encontró la especie (24 sitios) y en el eje Y el peso total (suma de los pesos). Además deben estar los datos apilados por sexo --> La gráfica debe mostrar el peso total por sexo para cada sitio.

````python
#Agrupemos los datos por sitio y por sexo
by_site_sex = surveys_df.groupby(['plot_id','sex'])
by_site_sex.head()
````

````python
# Luego calculamos la suma de los pesos para cada sexo por sitio
site_sex_count = by_site_sex['weight'].sum()
site_sex_count
````

### Se puede usar `.unstack()` en los datos agrupados para entender como el peso total de cada sexo contribuye a cada sitio.

````python
by_site_sex = surveys_df.groupby(['plot_id','sex'])
site_sex_count = by_site_sex['weight'].sum()
site_sex_count.unstack()
````

````python
##Grafica sexo por sitio
#Otros estilos https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

by_site_sex = surveys_df.groupby(['plot_id','sex'])
site_sex_count = by_site_sex['weight'].sum()
spc = site_sex_count.unstack()



# Grafica de barras
import matplotlib.pyplot as plt

#Otros estilos https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
plt.style.use("seaborn")            #Declaración del estilo

#grafico rápido con site_sex_count que ya lo calculamos antes
s_plot = spc.plot(kind='bar',legend=True,stacked=True,fontsize=16,figsize=(10,8))

s_plot.set_ylabel("Peso", fontsize=18)
s_plot.set_xlabel("Sitio_ID",fontsize=18)
s_plot.set_title("Total peso por sitio y sexo", fontsize=18)
````
